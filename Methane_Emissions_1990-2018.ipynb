```python
import pandas as pd
df = pd.read_csv('Methane_Emissions_1990-2018.csv')
```

```python
df_c[df_c.isna().any(axis=1)]
```

```python
df_c.query('Country in ["Micronesia", "Marshall Islands", "Palau"] \
                and Sector in ["Total excluding LUCF","Agriculture", "Land-Use Change and Forestry"]')[["Country","1990", "1991", "1992", "1993", "1994"]]
```

```python
df_c.loc[:,"1990"].fillna(0.0, inplace=True)
```

```python
df_c[df_c.isna().any(axis=1)]
```

```python
df_c = df_c.melt(id_vars=["Country", "Sector", "Gas", "Unit"], value_vars=[str(y) for y in range(1990,2019)], var_name="Year", value_name="Emission")
```

```python
df_c.query('Country == "India"').head(10)
```

```python
df_c = df_c.set_index(['Country', 'Year', 'Sector'])['Emission'].unstack(["Sector"]).reset_index()
df_c.columns.name= None
```

```python
df_c.query('Country=="Maldives" and Year < "1995"')
```

```python
df_c.fillna(0.0, inplace=True)
```

```python
df_c.isna().sum()
```

```python
df_c.to_csv('Methane_Emissions_1990-2018_cleaned.csv', index=False)
```
