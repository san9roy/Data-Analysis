```python
import pandas as pd
df = pd.read_csv("Student Mental health.csv")
```

```python
df_original = df
df_c = df.copy()
```

```python
df_c.rename(columns = {'date':'date' , 'Time':'time', 'Choose your gender':'gender',
                     'Age':'age', 'What is your course?':'course',
                     'Your current year of Study':'year_of_study',
                     'What is your CGPA?':'cgpa', 'Marital status': 'maritial_status',
                     'Do you have Depression?': 'depression', 'Do you have Anxiety?': 'anxeity',
                     'Do you have Panic attack?':'panic_attack',
                     'Did you seek any specialist for a treatment?':'treatment'} , inplace = True)
new_column_names = []
for i in df_c.columns:
   new_column_names.append(i.title())
df_c.columns = new_column_names
df_c.columns
```

```python
df_c['Course'] = df_c['Course'].str.title()
df_c.head(5)
```

```python
df_c[['L_E_CGPA', 'H_E_CGPA']] = df_c['Cgpa'].str.split('-', expand=True)
df_c = df_c.drop(columns = ["Cgpa"])
df_c.head(5)
```

```python
df_c["Age"] = df_c["Age"].fillna(0)
df_c["Age"] = df_c["Age"].astype(int)
df_c["Age"].dtype
```

```python
df_c["Year_Number"] = df_c["Year_Of_Study"].str.split(" ").str[1].astype(int)
df_c = df_c.drop(columns = ["Year_Of_Study"])
df_c.head(5)
```

```python
df_c[["L_E_CGPA", "H_E_CGPA"]] = df_c[["L_E_CGPA", "H_E_CGPA"]].astype(float)
df_c.dtypes
```
